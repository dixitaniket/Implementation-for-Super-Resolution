```python
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Conv2D,Concatenate,Input,UpSampling2D
import numpy as np
from tensorflow.keras.utils import plot_model
```

```python
# rdeep residual block (concept taken from the paper of Residual Dense Network for Image Super-Resolution)
def residual_block(filters,input_layer):
    # input_layer=Input(shape=input_shape)
    layers=[input_layer,]
    for i in range(1,20):
        if i==1:
            layers.append(Conv2D(filters,(3,3),activation="relu",padding="same",name=f"{i}")(layers[i-1]))
        elif i==2:
            layers.append(Conv2D(filters,(3,3),activation="relu",padding="same",name=f"{i}")(Concatenate()([layers[i-1],layers[i-2]])))
        else:
            layers.append(Conv2D(filters, (3, 3), activation="relu", padding="same", name=f"{i}")(Concatenate()(layers[1:])))

    layers.append(Conv2D(filters,(3,3),activation="relu",padding="same",name="last_layer")(Concatenate()(layers[1:])))
    layers.append(UpSampling2D(size=(2,2))(layers[-1]))
    layers.append(Conv2D(1,(3,3),activation="sigmoid",padding="same")(layers[-1]))
    return Model(input_layer,layers[-1])
```

```python
(x_train,_),(_,_)=tf.keras.datasets.mnist.load_data()
```

```python
x_train_lower=x_train
```

```python
import skimage 
from skimage.transform import downscale_local_mean
```

```python
import matplotlib.pyplot as plt
```

```python
data=downscale_local_mean(x_train,(1,2,2))
```

```python
plt.imshow(data[10],cmap="gray")
```

```python
plt.imshow(x_train[10],cmap="gray")
```

```python
data=data.reshape((-1,*data.shape[1:],1))
```

```python
# defining the input layer
input_layer=Input(shape=data.shape[1:])
```

```python
# defining the model
model=residual_block(5,input_layer)
```

```python
model.layers[-1]
```

```python
# seeing data shape
data.shape
```

```python
# plotting the model
plot_model(model,"someshape.png")
```

```python
model.summary()
```

```python
# compiling the model
model.compile(optimizer="adam",loss="binary_crossentropy",metrics=["accuracy"])
```

```python
# data
data=data/255
x_train=x_train/255
```

```python
x_train.shape
```

```python
data.shape
```

```python
# training the model
model.fit(data,x_train,epochs=30,batch_size=64)
```

```python
# displaying the result
plt.imshow(data[100].reshape(14,14),cmap="gray")
```

```python
ans=model.predict(data[100].reshape(1,14,14,1))
```

```python
plt.imshow(ans.reshape(28,28),cmap="gray")
```
